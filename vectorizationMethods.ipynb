## Bag-of-words

Q. Find Dissmilarity between X and Y,
X = [3,2,0,5]
Y = [1,0,0,0]
using Cosine similarity

```python
import numpy as np
X = np.array([3,2,0,5])
Y = np.array([1,0,0,0])

dot_X_Y = np.dot(X,Y)
mag_X = np.linalg.norm(X)
mag_Y = np.linalg.norm(Y)

Similarity = dot_X_Y / (mag_X * mag_Y)
print('Similarity: {}'.format(Similarity))
print('Dissimilarity: {}'.format(1-Similarity))
```

## TF-IDF

Q.
D1 = The cat sat on the mat\
D2 = The dog played in the park\
D3 = Cats and Dogs are great pets\
Calculate TF-IDF score for 'Cat'

```python
cat_d1 = 1/6
cat_d2 = 0/6
cat_d3 = 1/6
Tf = np.array([cat_d1, cat_d2, cat_d3])
print(f'Tf = {Tf}')
Idf = np.log10(3/2)
print(f'Idf = {Idf}')
Tfidf = Tf * Idf
print(f'Tfidf = {Tfidf}')
```
